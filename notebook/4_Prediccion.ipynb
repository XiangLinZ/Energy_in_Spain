```python
import pandas as pd
import pickle
```

```python
predicción = pd.read_csv("../data/visualizacion/predicción.csv", index_col = 0)
with open(f"../data/visualizacion/px_demanda_real.pkl", "rb") as demanda_real:
    demanda = pickle.load(demanda_real)
```

```python
predicción_demanda = predicción.reset_index().rename(columns = {"index": "Fecha"})
predicción_demanda["Fecha"] = pd.to_datetime(predicción_demanda["Fecha"], utc = True, format = "%Y-%m-%d %H:%M:%S%z")
predicción_demanda["Fecha"] = predicción_demanda["Fecha"].dt.tz_convert("Europe/Paris")
predicción_demanda.rename(columns = {"y_pred": "Prediccion propia en MW"}, inplace = True)
predicción_demanda
```

```python
demanda["drop"] = demanda["Fecha"].where(demanda["Fecha"].dt.minute % 10 == 0, other = "drop")
demanda = demanda[demanda["drop"] != "drop"]
demanda = demanda[["Fecha", "Demanda real en MW", "Demanda prevista en MW"]]
inicio = "2023-03-31 14:30:00+02:00"
final = "2023-03-31 18:20:00+02:00"
df_demanda = demanda[(demanda["Fecha"] >= inicio) & (demanda["Fecha"] <= final)]
df_demanda
```

```python
df_demanda_predicción = pd.merge(predicción_demanda,df_demanda)
df_demanda_predicción
```

```python
with open("../data/visualizacion/px_prediccion.pkl", "wb") as prediccion_demanda:
    pickle.dump(df_demanda_predicción, prediccion_demanda)
```
